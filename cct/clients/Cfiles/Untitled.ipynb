```python
cfile_name = '3d_trap14_cfile.csv'
cfile_text = open(cfile_name).read().replace('\r', '').split('\n')[:-1]
```

```python
cfile_text
```

```python
import csv

with open('3d_trap14_cfile_format.csv', 'w') as csvfile:
    cfile = csv.writer(csvfile, delimiter = ' ', escapechar = ' ', quoting = csv.QUOTE_NONE)
    for item in cfile_text:
        cfile.writerow(item)
```

```python
import numpy as np
np.savetxt('3d_trap14_cfile_format.csv', cfile_text, delimiter = ' ', newline = '\n')
```
